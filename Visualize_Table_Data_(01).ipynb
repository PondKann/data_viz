```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

## Parallel Coordinates
แสดงตัวเลขที่แบ่งเป็น class  เอา column มาต่อกันในแกน X

ข้อเสีย คือ ถ้ามีข้อมูลมาก  มีหลาย class มันจะมั่ว รวมกันไปหมด

```
pd.plotting.parallel_coordinates(df,'Name')
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#A506CE', '#F9B41E', '#F50C0C'))  #สีแรก คือ สีของ class บนสุด
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('y', 'm', 'c'))     
 #แม่สี ตัวย่อ R G B C M Y K  (red green blue), Cyan (ฟ้า) Magenta (แดงอมม่วง) Yellow (เหลือง) Key (สีดำ)
 
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.5,0), (0,0.8,0), (0.7,0,1)))
      #สามารถกำหนดปริมาณของแสงได้ ซึ่ง RGB มีค่าตั้งแต่ 0-1
```

ชุดสี colormap 
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```
pd.plotting.parallel_coordinates(df, 'Name', colormap = 'tab20b' ) 
```

# เริ่ม Matplotlip
library พื้นฐานสำหรับวาดกราฟต่าง ๆ 

```
from matplotlib import pyplot as plt   #from เป็นที่รวมแพกเก็จใหญ่  import มาแค่ pyplot
```

## แปลงข้อมูลจาก pandas เป็น numpy
- pandas มีหัวตาราง มีตัวเลขหรือตัวหนังสือข้างใน
- numpy มองตัวเลขเป็น matrix มีแค่ตัวเลข

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

# Visualize array data ด้วย pixel

แทนค่าข้างในด้วยจุดสี ค่ามาก สีจะสว่าง

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15]   #ขนาดของfigure (กว้าง*ยาว) 
plt.imshow(np_data[:10,:],interpolation='nearest')
```

```
plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
#column 2-3 จะมีขนาดเล็กกว่า 50 ตัวแรก จึงทำให้แยกได้ง่าย
plt.imshow(np_data[50:100,:],interpolation='nearest')  
```

### การใช้ subplot
นำกราฟมาเรียงต่อกัน

```
plt.subplot(1,3,1)          #subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่ 1,2)
plt.imshow(np_data[:50,:],interpolation='nearest', cmap='magma')
plt.subplot(1,3,2) 
plt.imshow(np_data[50:100,:],interpolation='nearest', cmap='magma')  
plt.subplot(1,3,3) 
plt.imshow(np_data[100:150,:],interpolation='nearest', cmap='magma')     #ดอกที่2 ค่าแรกจะมากสุด
```

```
plt.imshow(np_data[:10,:],interpolation='nearest', cmap='magma' , alpha=0.8)  #alpha ความโปร่ง
```
